## 2. TF-IDF

```python
vectorizer = TfidfVectorizer(stop_words= "english")
vectors = vectorizer.fit_transform(df["ENServiceName"])
feature_names = vectorizer.get_feature_names_out()
dense = vectors.todense()
denselist = dense.tolist()
df_dense_list = pd.DataFrame(denselist, columns=feature_names)
```

```python
df_dense_list.head(10)
```

```python
np.where(cosine_similarity(df_dense_list[1:2],df_dense_list[2:len(df_dense_list)-1]) == min(cosine_similarity(df_dense_list[1:2],df_dense_list[2:len(df_dense_list)-1])))
```

```python
dense
```

```python
#matrix = pd.DataFrame(cosine_similarity(denselist, denselist), index=denselist, columns=feature_names)
```

```python
lst = []
for index, row in df_dense_list.iterrows():
  lst.append(cosine_similarity(df_dense_list[1:2], df_dense_list[2:len(df_dense_list)]))
```
